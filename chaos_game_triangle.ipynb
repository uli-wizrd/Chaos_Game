# The chaos game

The chaos game is a simple dynamic  where:

1. You choose 3 points in a plane so they form a triangle it doesn't matter what kind they just have to form one, once we have our vertices A, B and C we can join them to form the boundary of our game board. 

2. Once we have our triangle we can choose a point at random as our starting position.

3. Now we assign each vertex of our triangle a pair of values from a 6 sided dice where each side has a dot in it (e.g A has the sides with 1 and 2 dots, B 3 dots and 4 dots and C 5 dots and 6 dots) once we do this we can roll our dice.

4. Depending of the face up value of our dice we will choose a point half way between our starting point and the vertex assigned to that value on our dice, now our starting point will be this new location

5. we repeat step 4 until something cool appears.




```python
# first some libraries as always

import matplotlib.pyplot as plt
import numpy as np
import math as mt
```

```python
# distance value

d = 0.5

#d = 0.15

#d = 0.61

# d = 0.75

# d = 0.43
```

```python
# triangle

tri = [[65,105,25,65],[120,20,20,120]]
```

```python
# initial point
i_px = 2
i_py = 5

# we define where we will store stuff

p_x = []
p_y = []

# we store our initial values
p_x.append(i_px)
p_y.append(i_py)
```

```python
# unrestricted triangle loop with a fixed distance of 0.5

np.random.seed(13)

for i in range(3000):
    
    dice_r = np.random.randint(1,4,dtype=int)

    if dice_r == 1:
            
            s_x = round(tri[0][0] + i_px, 4)
            s_y = round(tri[1][0] + i_py, 4)

            c_x = round((s_x*d), 4)
            c_y = round((s_y*d), 4)

            p_x.append(c_x)
            p_y.append(c_y)

            i_px = c_x
            i_py = c_y

    elif dice_r == 2:

            s_x = round(tri[0][1] + i_px, 4)
            s_y = round(tri[1][1] + i_py, 4)

            c_x = round((s_x*d), 4)
            c_y = round((s_y*d), 4)

            p_x.append(c_x)
            p_y.append(c_y)

            i_px = c_x
            i_py = c_y

    elif dice_r == 3:

            s_x = round(i_px + tri[0][2], 4)
            s_y = round(i_py + tri[1][2], 4)

            c_x = round((s_x*d), 4)
            c_y = round((s_y*d), 4)

            p_x.append(c_x)
            p_y.append(c_y)

            i_px = c_x
            i_py = c_y
```

```python
# restricted triangle loop with fixed distance

np.random.seed(67)

pre_roll = np.random.randint(1,4,dtype=int)

for i in range(8000):
    
    dice_r = np.random.randint(1,4,dtype=int)
    ctrl = 0

    if dice_r == pre_roll:

        while dice_r == pre_roll:

            ctrl += 1
            dice_r = np.random.randint(1,4,dtype=int)

            if dice_r != pre_roll:
                break
            elif ctrl > 10:
                break
            else:
                pass
    else:
        pass

    pre_roll = dice_r

    if dice_r == 1:
            
            s_x = round(tri[0][0] + i_px, 4)
            s_y = round(tri[1][0] + i_py, 4)

            c_x = round((s_x*d), 4)
            c_y = round((s_y*d), 4)

            p_x.append(c_x)
            p_y.append(c_y)

            i_px = c_x
            i_py = c_y

    elif dice_r == 2:

            s_x = round(tri[0][1] + i_px, 4)
            s_y = round(tri[1][1] + i_py, 4)

            c_x = round((s_x*d), 4)
            c_y = round((s_y*d), 4)

            p_x.append(c_x)
            p_y.append(c_y)

            i_px = c_x
            i_py = c_y

    elif dice_r == 3:

            s_x = round(i_px + tri[0][2], 4)
            s_y = round(i_py + tri[1][2], 4)

            c_x = round((s_x*d), 4)
            c_y = round((s_y*d), 4)

            p_x.append(c_x)
            p_y.append(c_y)

            i_px = c_x
            i_py = c_y
```

```python
# transfer function triangle loop
# we will space our points in term of their distance to the other point
# we will use the sigmoid function

def fun (x):
    
    #y = 1/(1 + pow(mt.e,-x))
    #y = mt.tanh(x)
    y = mt.sin(2*x)

    return y

np.random.seed(41)

tc = 0.09

# dynamic distance between points

for i in range(8000):
    
    dice_r = np.random.randint(1,4,dtype=int)

    if dice_r == 1:
            
            s_x = round(tri[0][0] + i_px, 4)
            s_y = round(tri[1][0] + i_py, 4)

            d = ( tri[0][0] - i_px )**2 + ( tri[1][0] - i_py )**2

            c_x = round((s_x * fun(d) * tc), 4)
            c_y = round((s_y * fun(d) * tc), 4)

            p_x.append(c_x)
            p_y.append(c_y)

            i_px = c_x
            i_py = c_y

    elif dice_r == 2:

            s_x = round(tri[0][1] + i_px, 4)
            s_y = round(tri[1][1] + i_py, 4)

            d = ( tri[0][1] - i_px )**2 + ( tri[1][1] - i_py )**2

            c_x = round((s_x * fun(d) * tc), 4)
            c_y = round((s_y * fun(d) * tc), 4)

            p_x.append(c_x)
            p_y.append(c_y)

            i_px = c_x
            i_py = c_y

    elif dice_r == 3:

            s_x = round(i_px + tri[0][2], 4)
            s_y = round(i_py + tri[1][2], 4)

            d = ( tri[0][2] - i_px )**2 + ( tri[1][2] - i_py )**2

            c_x = round((s_x * fun(d) * tc), 4)
            c_y = round((s_y * fun(d) * tc), 4)

            p_x.append(c_x)
            p_y.append(c_y)

            i_px = c_x
            i_py = c_y
```

```python
# transforming results to plot them
p_x = np.array(p_x).reshape(-1,1)
p_y = np.array(p_y).reshape(-1,1)
dots = np.hstack((p_x,p_y))
```

```python
# triangle graph

fig, ax = plt.subplots(facecolor = 'black' , figsize=(7,7))
ax.set_title("Sierpinski Gasket",color = 'w' )
#plt.plot(tri[0],tri[1], 'o-k')
plt.scatter(dots[:,0],dots[:,1],s=1, c = dots[:,1], cmap = plt.get_cmap("plasma"),
            alpha = 0.9)
plt.axis('off')
plt.show()
plt.close
```
