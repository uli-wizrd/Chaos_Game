# Variation on the chaos game

In this variation of the chaos game a 5 sided board is used

```python
# first some libraries as always

import matplotlib.pyplot as plt
import numpy as np
```

```python
# distance value

d=0.15

#d=0.75

#d=0.25
```

```python
# initial point
i_px = 35
i_py = 35
```

```python
# pentagon frame

pent = [[0,200,200,100,0,0],[0,0,100,150,100,0]]
```

```python
# we define where we will store stuff

p_x = []
p_y = []
```

```python
# we store our initial values
p_x.append(i_px)
p_y.append(i_py)
```

```python
# pentagon loop unresstricted

np.random.seed(41)

for i in range(8000):
    
    dice_r = np.random.randint(1,6,dtype=int)

    if dice_r == 1:
            
            s_x = round(pent[0][0] + i_px, 4)
            s_y = round(pent[1][0] + i_py, 4)

            c_x = round((s_x*d), 4)
            c_y = round((s_y*d), 4)

            p_x.append(c_x)
            p_y.append(c_y)

            i_px = c_x
            i_py = c_y

    elif dice_r == 2:

            s_x = round(pent[0][1] + i_px, 4)
            s_y = round(pent[1][1] + i_py, 4)

            c_x = round((s_x*d), 4)
            c_y = round((s_y*d), 4)

            p_x.append(c_x)
            p_y.append(c_y)

            i_px = c_x
            i_py = c_y

    elif dice_r == 3:

            s_x = round(i_px + pent[0][2], 4)
            s_y = round(i_py + pent[1][2], 4)

            c_x = round((s_x*d), 4)
            c_y = round((s_y*d), 4)

            p_x.append(c_x)
            p_y.append(c_y)

            i_px = c_x
            i_py = c_y

    elif dice_r == 4:

            s_x = round(i_px + pent[0][3], 4)
            s_y = round(i_py + pent[1][3], 4)

            c_x = round((s_x*d), 4)
            c_y = round((s_y*d), 4)

            p_x.append(c_x)
            p_y.append(c_y)

            i_px = c_x
            i_py = c_y

    elif dice_r == 5:

            s_x = round(i_px + pent[0][4], 4)
            s_y = round(i_py + pent[1][4], 4)

            c_x = round((s_x*d), 4)
            c_y = round((s_y*d), 4)

            p_x.append(c_x)
            p_y.append(c_y)

            i_px = c_x
            i_py = c_y
```

```python
# chaps game loop for the pentagon with restrictions

np.random.seed(67)

pre_roll = np.random.randint(1,6,dtype=int)

for i in range(12000):
    
    dice_r = np.random.randint(1,6,dtype=int)
    ctrl = 0

    if dice_r == pre_roll:

        while dice_r == pre_roll:

            ctrl += 1
            dice_r = np.random.randint(1,6,dtype=int)

            if dice_r != pre_roll:
                break
            elif ctrl > 10:
                break
            else:
                pass
    else:
        pass

    pre_roll = dice_r

    if dice_r == 1:
            
            s_x = round(pent[0][0] + i_px, 4)
            s_y = round(pent[1][0] + i_py, 4)

            c_x = round((s_x*d), 4)
            c_y = round((s_y*d), 4)

            p_x.append(c_x)
            p_y.append(c_y)

            i_px = c_x
            i_py = c_y

    elif dice_r == 2:

            s_x = round(pent[0][1] + i_px, 4)
            s_y = round(pent[1][1] + i_py, 4)

            c_x = round((s_x*d), 4)
            c_y = round((s_y*d), 4)

            p_x.append(c_x)
            p_y.append(c_y)

            i_px = c_x
            i_py = c_y

    elif dice_r == 3:

            s_x = round(i_px + pent[0][2], 4)
            s_y = round(i_py + pent[1][2], 4)

            c_x = round((s_x*d), 4)
            c_y = round((s_y*d), 4)

            p_x.append(c_x)
            p_y.append(c_y)

            i_px = c_x
            i_py = c_y

    elif dice_r == 4:

            s_x = round(i_px + pent[0][3], 4)
            s_y = round(i_py + pent[1][3], 4)

            c_x = round((s_x*d), 4)
            c_y = round((s_y*d), 4)

            p_x.append(c_x)
            p_y.append(c_y)

            i_px = c_x
            i_py = c_y

    elif dice_r == 5:

            s_x = round(i_px + pent[0][4], 4)
            s_y = round(i_py + pent[1][4], 4)

            c_x = round((s_x*d), 4)
            c_y = round((s_y*d), 4)

            p_x.append(c_x)
            p_y.append(c_y)

            i_px = c_x
            i_py = c_y
```
